```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sal_1 = np.array([100,200,400,300])
sal_2 = np.array([50,300,500,450])
```

```python
# add Label
plt.plot(sal_1, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(sal_2, c='Red', ls='--', marker='s', ms=8, label='Salários Gileno')
plt.show()
```

```python
# add Legenda
plt.plot(sal_1, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(sal_2, c='Red', ls='--', marker='^', ms=8, label='Salários Gileno')
plt.legend()
plt.show()
```

```python
plt.plot(sal_1, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(sal_2, c='Red', ls='--', marker='^', ms=8, label='Salários Gileno')
plt.legend(loc='upper left')
plt.show()
```

```python
# add line
sal_3 = [200,150,300,700]
plt.plot(sal_1, c='Black', ls='--', marker='s', ms=10, label='Salários Marcos')
plt.plot(sal_2, c='Red', ls='--', marker='^', ms=10, label='Salários Gileno')
plt.plot(sal_3, c='Green', ls='--', marker='o', ms=10, label='Salários Pedro')
plt.legend(loc='upper left')
plt.show()
```
