```python
class Point(object):
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
class Reward(Point):
    
    def __init__(self, x, y, name):
        super(Reward, self).__init__(x,y)
        self.name = name
        
class Robo2D(Point):
           
    def move_up(self):
        if self.y < 10:
            self.y = self.y + 1
        else:
            print('Movimento proibido')
            
    def move_down(self):
        if self.y > 0:
            self.y = self.y - 1
        else:
            print('Movimento proibido')
            
    def move_rigth(self):
        if self.x > 10:
            self.x = self.x + 1
        else:
            print('Movimento proibido')
        
    def move_left(self):
        if self.x > 0:
            self.x = self.x - 1
        else:
            print('Movimento proibido')
```

```python
def check_reward(robot, rewards):
    ok = False
    for reward in rewards:
        if reward.x == robot.x and reward.y == robot.y:
            print("O robô achou a recompensa: %s" % reward.name)
            ok = True
    return ok
```

```python
r1 = Reward(1, 2, 'Moeda')
r2 = Reward(5, 3, 'gasolina')

robot = Robo2D(2,2)
rewards = [r1, r2]
check_reward(robot, rewards)
robot.move_left()
print(robot.x)
print(robot.y)
check_reward(robot, rewards)
```
